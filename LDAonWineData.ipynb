```python
from sklearn.datasets import load_wine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=load_wine()
x=np.array(wine.data) #178X13 array --- input feature vectors
y=np.array(wine.target) #178 array --- output labels
print(x[1:5,:])
print(y)
```

```python
wine
```

```python
#Applying PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
print(x_pca)
plt.scatter(x[:,0],x[:,1],c=y)
```

Thereis mixed clustering in PCA

```python
#Apply LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=2)
x_lda=lda.fit_transform(x,y)
print(x_lda)
plt.scatter(x_lda[:,0],x_lda[:,1],c=y)
```

Now data is linearly seperable,and it is reduced into 2 classes from originally 13 classes

Now LDA can be used as linear classifier also.... We will divide our data into 2 parts. One for training and other for testing

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x,y,test_size=0.3) #training data is 70% and testing is 30%
```

```python
lda.fit(xtrain,ytrain) #now training is over
```

```python
#now we will go for prediction on testing data
yn=lda.predict(xtest) #now this yn is predicted class wich was originally ytest
```

```python
print(ytest) #true classes
print(yn) #predicted class
print(ytest-yn) #difference between these 2
```
