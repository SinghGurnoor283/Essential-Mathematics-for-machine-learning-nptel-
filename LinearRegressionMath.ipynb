```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
score = pd.DataFrame(np.array([[2.5 ,21] ,[5.1 , 47], [3.2, 27] ,[8.5,75] , [3.5 , 50], [1.5, 20]]),

                         columns = ["Hours" , "Scores"])
```

```python
score
```

```python
score.describe()
```

```python
score.plot(x='Hours',y='Scores',style='o')
plt.show()
```

```python
X=score.iloc[:,:-1].values; #ist column(hours), it is independent variable
y=score.iloc[:,1].values #2nd column. because ist coulmn is zero... dependent variable
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression() #object
reg.fit(X,y)
```

```python
print(reg.intercept_) #it gives a0
```

```python
print(reg.coef_) # it is a1. y=7.2821 + 8.0784x
```

```python
y_pred = reg.predict(X) #find new value of y (predicts)
```

```python
df=pd.DataFrame({'Actual':y,'Predicted':y_pred})
df
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('MSE: %.2f' % mean_squared_error(y,y_pred))
```

```python
print('R2 score: %.2f' % r2_score(y,y_pred))
```

```python
plt.scatter(X,y, color='black')
plt.plot(X,y_pred,color='red', linewidth=2)
plt.show()
```
